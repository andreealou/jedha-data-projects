## Coordinates for all the cities

```

cities = ["Mont Saint Michel", "St Malo", "Bayeux", "Le Havre", "Rouen", "Paris", "Amiens", "Lille", "Strasbourg",
    "Chateau du Haut Koenigsbourg", "Colmar", "Eguisheim", "Besancon", "Dijon", "Annecy", "Grenoble",
    "Lyon", "Gorges du Verdon", "Bormes les Mimosas", "Cassis", "Marseille", "Aix en Provence",
    "Avignon", "Uzes", "Nimes", "Aigues Mortes", "Saintes Maries de la mer", "Collioure", "Carcassonne",
    "Ariege", "Toulouse", "Montauban", "Biarritz", "Bayonne", "La Rochelle"]

data = []

for city in cities:
    try:
        params = {
            "q": f"{city}, France",
            "format": "json",
            "limit": 1
        }
        
        headers =  {"User-Agent": "MyPythonApp/1.0"}
        
        response = requests.get(nominatim_url, params=params, headers= headers)
                
        results = response.json()
        
        if results:
            lat = results[0]["lat"]
            lon = results[0]["lon"]
            print(f"{city}: {lat}, {lon}")
            data.append({"Ville": city, "Latitude": lat, "Longitude": lon})
        else:
            print(f"Aucune donnée trouvée pour {city}")
            data.append({"Ville": city, "Latitude": None, "Longitude": None})

    except Exception as e:
        print(f"Erreur pour {city}: {e}")
        data.append({"Ville": city, "Latitude": None, "Longitude": None})

# Création d'un DataFrame
df = pd.DataFrame(data)

# Sauvegarde en CSV
df.to_csv("coordonnees_villes.csv", index=False)
```

```
import requests
import json
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os
```

```
# Chargement de la clé de l'API depuis .env
# J'ai conscience que le but de .env est de ne pas révéler la clé mais j'ai eu des problèmes pour la charger et cela m'a permis de vérifier sa véracité

load_dotenv()

OW_KEY = os.getenv("OW_KEY")

if OW_KEY is None:
    print("Erreur Clé API")
else:
    print("Clé API chargée:", OW_KEY)
```

```
# Chargement du fichier CSV des villes et leurs coordonnées GPS
df_villes = pd.read_csv("coordonnees_villes.csv")

# Vérification du data frame 
print(df_villes)
```

```
# Liste pour stocker les résultats
resultats = []

# Boucle sur chaque ville
for index, row in df_villes.iterrows():
    ville = row["Ville"]
    lat = row["Latitude"]
    lon = row["Longitude"]
    
    ow_5D_forecast_url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={OW_KEY}&units=metric"
    response = requests.get(ow_5D_forecast_url)
    data = response.json()

    if response.status_code == 200 and "list" in data:
        temperatures = []
        humidites = []
        vents = []
        couverture_nuageuse = []
        conditions = []
        
        # Parcourir les prévisions horaires sur 5 jours (toutes les 3 heures)
        for item in data["list"]:
            temp = item["main"]["temp"]
            humidity = item["main"]["humidity"]
            wind_speed = item["wind"]["speed"]
            cloudiness = item["clouds"]["all"]
            weather_main = item["weather"][0]["main"]  # Type de météo (Clear, Rain, Clouds, etc.)
            
            temperatures.append(temp)
            humidites.append(humidity)
            vents.append(wind_speed)
            couverture_nuageuse.append(cloudiness)
            # Attribuer une valeur numérique à chaque condition
            if weather_main == "Clear":
                conditions.append(0.1)  # Beau temps
            elif weather_main == "Rain":
                conditions.append(1)  # Pluie
            elif weather_main == "Clouds":
                conditions.append(0.6)  # Nuages
            elif weather_main == "Snow":
                conditions.append(1)  # Neige (valeur à ajuster selon vos préférences)
            elif weather_main == "Mist":
                conditions.append(1)  # Brouillard (valeur à ajuster)
                                       
        
        # Calculer la moyenne des 5 jours
        temp_moyenne = sum(temperatures) / len(temperatures)
        humidite_moyenne = sum(humidites) / len(humidites)
        vent_moyen = sum(vents) / len(vents)
        couverture_nuageuse_moyenne = sum(couverture_nuageuse) / len(couverture_nuageuse)
        condition_moyenne = sum(conditions) / len(conditions)
        if condition_moyenne < 0.21:
            condition_text = "Le temps est majoritairement beau sur 5 jours."
        elif condition_moyenne < 0.45:
            condition_text = "Le temps est généralement beau avec quelques nuages."
        elif condition_moyenne < 0.8:
            condition_text = "Le temps est partiellement nuageux."
        else:
            condition_text = "Le temps sera majoritairement pluvieux ou nuageux."
        
        print(f"Moyennes meteo pour {ville} sur 5 jours :")
        print(f"Temperature moyenne : {temp_moyenne:.1f}°C")
        print(f"Humidite moyenne : {humidite_moyenne:.1f}%")
        print(f"Vent moyen : {vent_moyen:.1f} m/s")
        print(f"Couverture nuageuse moyenne : {couverture_nuageuse_moyenne:.1f}%")
        print(f"Conditions Meteo : {condition_text}\n")
       
        
        resultats.append({
            "Ville": ville,
            "Latitude": lat,
            "Longitude": lon,
            "Temperature Moy (°C)": round(temp_moyenne, 1),
            "Humidite Moy (%)": round(humidite_moyenne, 1),
            "Vent Moy (m/s)": round(vent_moyen, 1),
            "Couverture nuageuse (%)": round(couverture_nuageuse_moyenne, 1),
            "Conditions Meteo": condition_text,
        })
    else:
        print(f"Aucune donnee pour {ville}")

# Convertir en DataFrame et sauvegarder
df_resultats = pd.DataFrame(resultats)
df_resultats.to_csv("moyenne_meteo_5jours.csv", index=False, encoding="utf-8")
```

## Trier mon dataset par meilleure meteo

```
# Normaliser les valeurs (mettre tout entre 0 et 1 pour comparer)
scaler = MinMaxScaler()

dataset_sorted["temp_norm"] = scaler.fit_transform(dataset_sorted[["Temperature Moy (°C)"]])
dataset_sorted["vent_norm"] = 1 - scaler.fit_transform(dataset_sorted[["Vent Moy (m/s)"]])  # Inverser car moins = mieux
dataset_sorted["nuage_norm"] = 1 - scaler.fit_transform(dataset_sorted[["Couverture nuageuse (%)"]])  # Inverser

# Attribuer un score aux conditions météo
condition_scores = {
    "Ensoleillé": 1.0,
    "Partiellement nuageux": 0.7,
    "Nuageux": 0.5,
    "Pluvieux": 0.3,
    "Orageux": 0.1
}
dataset_sorted["conditions_score"] = dataset_sorted["Conditions Meteo"].map(condition_scores)

# Calculer un score final pondéré
dataset_sorted["score_meteo"] = (
    0.1 * dataset_sorted["temp_norm"] + 
    0.4 * dataset_sorted["vent_norm"] + 
    0.1 * dataset_sorted["nuage_norm"] + 
    0.4 * dataset_sorted["conditions_score"]
)

# Trier selon le meilleur score
top_villes = dataset_sorted.sort_values(by="score_meteo", ascending=False)

# Afficher le classement
print(top_villes[["Ville", "Temperature Moy (°C)", "Vent Moy (m/s)", "Couverture nuageuse (%)", "Conditions Meteo", "score_meteo"]])
```

## Classement de chaque ville avec moyenne pondéré

```
# Calculer le classement pour chaque critère
df_sorted_temp = df_resultats.sort_values(by="Temperature Moy (°C)", ascending=False)
df_sorted_temp["rank_temp"] = range(1, len(df_sorted_temp) + 1)

df_sorted_humidity = df_resultats.sort_values(by="Humidite Moy (%)", ascending=True)  # Moins d'humidité est mieux
df_sorted_humidity["rank_humidity"] = range(1, len(df_sorted_humidity) + 1)

df_sorted_vent = df_resultats.sort_values(by="Vent Moy (m/s)", ascending=True)  # Moins de vent est mieux
df_sorted_vent["rank_vent"] = range(1, len(df_sorted_vent) + 1)

df_sorted_nuage = df_resultats.sort_values(by="Couverture nuageuse (%)", ascending=True)  # Moins de nuages est mieux
df_sorted_nuage["rank_nuage"] = range(1, len(df_sorted_nuage) + 1)

# Fusionner les DataFrames avec les classements
df_sorted_all = df_sorted_temp[["Ville", "rank_temp"]].merge(df_sorted_humidity[["Ville", "rank_humidity"]], on="Ville") \
    .merge(df_sorted_vent[["Ville", "rank_vent"]], on="Ville") \
    .merge(df_sorted_nuage[["Ville", "rank_nuage"]], on="Ville")

# Vérifier le DataFrame avec les classements
print(df_sorted_all.head())
```

```
# Poids pour chaque critère
poids = {
    "rank_temp": 0.3,
    "rank_humidity": 0.2,
    "rank_vent": 0.2,
    "rank_nuage": 0.3
}

# Calcul de la moyenne pondérée
df_sorted_all["score_final"] = (
    poids["rank_temp"] * df_sorted_all["rank_temp"] +
    poids["rank_humidity"] * df_sorted_all["rank_humidity"] +
    poids["rank_vent"] * df_sorted_all["rank_vent"] +
    poids["rank_nuage"] * df_sorted_all["rank_nuage"]
)

# Trie par le score final pour obtenir le classement
df_sorted_final = df_sorted_all.sort_values(by="score_final", ascending=True)  # Plus petit est mieux, donc on trie par score croissant

# Affichage les résultats finaux
print(df_sorted_final[["Ville", "score_final"]])

df_sorted_final.to_csv("classement_villes_pour_hotels.csv", index=False, encoding="utf-8")
```

```
import unicodedata
import numpy as np
import pandas as pd

# 0) Si tu n'as pas df_villes en mémoire, relis-le
# df_villes = pd.read_csv("coordonnees_villes.csv")

def normalize_name(s: pd.Series) -> pd.Series:
    # retire accents, espaces superflus et met en minuscule pour join robuste
    return (
        s.astype(str)
         .str.normalize('NFKD')
         .str.encode('ascii', errors='ignore')
         .str.decode('utf-8')
         .str.strip()
         .str.lower()
    )

# 1) Prépare des clés normalisées pour la jointure
df_sorted_final = df_sorted_final.copy()
df_sorted_final["__key__"] = normalize_name(df_sorted_final["Ville"])
df_villes_norm = df_villes.copy()
df_villes_norm["__key__"] = normalize_name(df_villes_norm["Ville"])

# 2) Merge pour récupérer Latitude/Longitude
df_sorted_final = df_sorted_final.merge(
    df_villes_norm[["__key__", "Latitude", "Longitude"]],
    on="__key__", how="left"
)

# 3) Nettoyage : enlève les villes sans coordonnées
manquantes = df_sorted_final[df_sorted_final[["Latitude","Longitude"]].isna().any(axis=1)]
if len(manquantes):
    print("Villes sans coordonnées (à corriger dans coordonnees_villes.csv) :")
    print(manquantes[["Ville"]].drop_duplicates().to_string(index=False))

df_clean = df_sorted_final.dropna(subset=["Latitude","Longitude"]).copy()

# 4) Prends les 5 meilleures exploitables (score plus petit = meilleur)
top5 = df_clean.nsmallest(5, "score_final").copy().reset_index(drop=True)
top5["Rank"] = top5.index + 1

# 5) Taille des bulles (plus petit score → plus grosse bulle)
eps = 1e-9
top5["size_inv"] = (top5["score_final"].max() - top5["score_final"]) + eps
```
